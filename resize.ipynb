### 이미지 해상도

```python
import os
from PIL import Image, ImageFile

# 손상된 파일도 강제로 읽도록 허용
ImageFile.LOAD_TRUNCATED_IMAGES = True

# 대상 폴더 경로
folder_path = "C:/Users/user/Downloads/VS_2.직접촬영_01.금속캔_002.알루미늄캔/232.재활용품_분류_및_선별_데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_2.직접촬영_01.금속캔_002.알루미늄캔_unzip"

# 줄이고 싶은 해상도 (일단 예시: 1024*576)
new_size = (1024, 576)

# 결과 저장할 폴더 경로
output_folder = os.path.join(folder_path, "resized")
os.makedirs(output_folder, exist_ok=True)

bad_files = [] # 문제 발생한 파일 기록용 

# 폴더 내 모든 jpg 파일 탐색
for index, filename in enumerate(os.listdir(folder_path)):
    if filename.lower().endswith(".jpg"):
        # 파일 경로
        file_path = os.path.join(folder_path, filename)
        # 저장 경로
        save_path = os.path.join(output_folder, filename)

        try:
            with Image.open(file_path) as img:
                # 해상도 줄이기
                resized_img = img.resize(new_size, Image.Resampling.LANCZOS)

                # 저장
                resized_img.save(save_path, "JPEG", quality=90)

                if(index % 100 == 0):
                    print(f"{index}번째 진행중")
        
        except Exception as e:
            print(f"error 오류 발생: {filename} → {e}")
            bad_files.append(filename)
        if index == 1: break

print("Resizing completed.")

if bad_files:
    print("문제가 발생한 파일 목록:", bad_files[:10], f"... 총 {len(bad_files)}개")


```

### 센터크롭으로 줄이기

```python
from PIL import Image, ImageFile
import os

# Allow truncated images to be loaded
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Target folder path
folder_path = "C:/Users/user/Downloads/VS_2.직접촬영_07.유리병_003.투명/232.재활용품_분류_및_선별_데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_2.직접촬영_07.유리병_003.투명_unzip"

# Target resolution
target_size = (512, 512)

# Output folder path
output_folder = os.path.join(folder_path, "resized_with_center_crop")
os.makedirs(output_folder, exist_ok=True)

bad_files = [] # For tracking problematic files

# Iterate through all jpg files in the folder
for index, filename in enumerate(os.listdir(folder_path)):
    if filename.lower().endswith(".jpg"):
        file_path = os.path.join(folder_path, filename)
        save_path = os.path.join(output_folder, filename)

        try:
            with Image.open(file_path) as img:
                original_width, original_height = img.size
                target_width, target_height = target_size

                # Resize the image so the shorter side matches the target size
                if original_width > original_height:
                    # Landscape image: Resize by height
                    new_height = target_height
                    new_width = int(original_width * (target_height / original_height))
                else:
                    # Portrait or square image: Resize by width
                    new_width = target_width
                    new_height = int(original_height * (target_width / original_width))
                
                img = img.resize((new_width, new_height), Image.Resampling.LANCZOS)
                
                # Calculate coordinates for center crop
                left = (new_width - target_width) / 2
                top = (new_height - target_height) / 2
                right = (new_width + target_width) / 2
                bottom = (new_height + target_height) / 2
                
                # Perform the center crop
                cropped_img = img.crop((left, top, right, bottom))
                
                # Save the final image
                cropped_img.save(save_path, "JPEG", quality=90)

                if index % 100 == 0:
                    print(f"Processing {index}th file: {filename}")
        
        except Exception as e:
            print(f"Error processing {filename} -> {e}")
            bad_files.append(filename)
        
        # Uncomment to test with a smaller number of files
        # if index == 1: break

print("Resizing and cropping completed.")

if bad_files:
    print(f"List of problematic files: {bad_files[:10]} ... Total {len(bad_files)} files.")
```

### 패딩넣어서 줄이기

```python
from PIL import Image, ImageFile
import os

# 손상된 파일도 강제로 읽도록 허용
ImageFile.LOAD_TRUNCATED_IMAGES = True

# 대상 폴더 경로
folder_path = "C:/Users/user/Downloads/VS_2.직접촬영_01.금속캔_002.알루미늄캔/232.재활용품_분류_및_선별_데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS_2.직접촬영_01.금속캔_002.알루미늄캔_unzip"

# 목표 해상도
target_size = (512, 512)

# 결과 저장할 폴더 경로
output_folder = os.path.join(folder_path, "resized_with_padding")
os.makedirs(output_folder, exist_ok=True)

bad_files = [] # 문제 발생한 파일 기록용

# 폴더 내 모든 jpg 파일 탐색
for index, filename in enumerate(os.listdir(folder_path)):
    if filename.lower().endswith(".jpg"):
        # 파일 경로
        file_path = os.path.join(folder_path, filename)
        # 저장 경로
        save_path = os.path.join(output_folder, filename)

        try:
            with Image.open(file_path) as img:
                original_width, original_height = img.size
                target_width, target_height = target_size

                # 원본 이미지의 종횡비 유지하며 크기 조정
                # 긴 변을 기준으로 리사이즈 비율 계산
                ratio = min(target_width / original_width, target_height / original_height)
                new_width = int(original_width * ratio)
                new_height = int(original_height * ratio)

                # 새로운 크기로 이미지 리사이즈
                resized_img = img.resize((new_width, new_height), Image.Resampling.LANCZOS)

                # 패딩을 추가할 새로운 이미지(캔버스) 생성
                padded_img = Image.new("RGB", target_size, (0, 0, 0)) # 검은색 패딩

                # 리사이즈된 이미지를 정중앙에 붙여넣기
                left_padding = (target_width - new_width) // 2
                top_padding = (target_height - new_height) // 2
                padded_img.paste(resized_img, (left_padding, top_padding))

                # 저장
                padded_img.save(save_path, "JPEG", quality=90)

                if(index % 100 == 0):
                    print(f"{index}번째 진행중: {filename}")
        
        except Exception as e:
            print(f"오류 발생: {filename} -> {e}")
            bad_files.append(filename)
        
        # 테스트를 위해 100개만 실행 후 종료하려면 아래 주석 해제
        #if index == 1: break

print("Resizing completed.")

if bad_files:
    print("문제가 발생한 파일 목록:", bad_files[:10], f"... 총 {len(bad_files)}개")
```
